## Handling High Sales Amount Values

#### In our dataset, we observe that some sales amounts are significantly higher than others. While these values might appear as outliers in a statistical sense, they are legitimate sales figures and represent actual transactions. Therefore, we cannot simply drop them from our analysis. To address this, we normalize the sales amount data using Min-Max scaling. This technique scales the data to a range of [0, 1], making it easier to compare and analyze without losing the integrity of the high sales values.


```python
#Code for Normalizing Sales Amount
from sklearn.preprocessing import MinMaxScaler 
scaler = MinMaxScaler() 
df['SalesAmount_Normalized'] = scaler.fit_transform(df[['SalesAmount']])
```

```python
fig, ax = plt.subplots(1,1,figsize=(4,4),gridspec_kw={'wspace': 0.3})
sns.boxplot(df.SalesAmount_Normalized, ax = ax)


plt.show()
```
